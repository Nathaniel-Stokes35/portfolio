<h2>Building the Models</h2>
<p>&nbsp;&nbsp;&nbsp;&nbsp;I'm going to be using a Logisitc Regression Model to try and predict the "product" column in each of the tables. I will be splitting each region into a 25:75 ratio for testing and training respectivly. I will be saving the predicted values and using them to determine the RMSE deviation of each model, then I will print the average volume of reserves.</p> 

```python
# Creating Models, Features, and Target Lists with matching indexs as the region list (For Ensambling)
models = [] # Storing the models for Cross-Validation
f_train = [] # Features_train
t_train = [] # Target_train
f_test = [] # Features_test
t_test = [] # Target_test
t_indices = [] # Test indicies
predictions = []

#Scaling and predicting with the model for each region, then Presenting the Score of Each
def scale_predict_RMSE(region): 
    print(f'\nRegion {idx}:\n') 
    
    # Assigning Features and Target varibles with the current DataFrame
    features = df[['f0', 'f1', 'f2']]
    target = df['product']
    
    # Train_Test_Split
    features_train, features_test, target_train, target_test = train_test_split(
        features, target, test_size=0.25, random_state=state
    )
    features_indices = features_test.index
    
    # Scaling those features and targets with Scaler
    features_train = scaler.fit_transform(features_train)
    features_test = scaler.transform(features_test)
    
    # Implementing Polynomial Features
    features_train = poly.fit_transform(features_train)
    features_test = poly.transform(features_test)
    
    # Assigning indexes and storing in lists
    f_train.append(features_train)
    t_train.append(target_train)
    f_test.append(features_test)
    t_test.append(target_test)
    t_indices.append(features_indices)
    
    # Creating the Model and making predictions
    LR_model = LinearRegression()
    LR_model.fit(features_train, target_train)
    models.append(LR_model)
    
    # Making Predictions with the fresh model
    predictions.append(LR_model.predict(features_test))
    region[idx].loc[features_indices, 'predicted_value'] = predictions[idx]
    
    # Testing those predictions and printing the RMSE
    rmse = np.sqrt(mean_squared_error(target_test, predictions[idx]))
    print(f'RMSE: {rmse}')
    
```

<h3>Improving RMSE for region 0 and region 2</h3>
<p>&nbsp;&nbsp;&nbsp;&nbsp;I'm going to be using bootstrapping and cross validation to find an average of the predictions among several models then I will use those averages for the RMSE, finding a more realistic measure for those more diverse regions.</p>
<h4>Bootstraping</h4>

```python
boot_itrs = 250 # Boot iterations

for idx in [0, 2]:
    
    # Setting Region variables
    predictions = []
    print(f'Region: {idx}...')
    
    for _ in range(boot_itrs): # Bootstrapping
        features_resampled, target_resampled = resample(
            f_train[idx], 
            t_train[idx], 
            random_state=state
        )
        
        # Constructing each Boot Model
        model = LinearRegression()
        model.fit(features_resampled, target_resampled)
        
        # Listing Predictions made from the Fresh Model
        predictions.append(model.predict(f_test[idx]))
    
    # Preforming and Printing Region Averages
    boot_avg = np.mean(predictions, axis=0)
    test_indices = t_indices[idx] 
    
    # Assiging indices for RMSE placement in the dataset
    region[idx].loc[test_indices, 'boot_prediction'] = boot_avg
    rmse = np.sqrt(mean_squared_error(t_test[idx], boot_avg))
    print(f'RMSE: {rmse}')
```

<h4>Cross-Validation</h4>

```python
for idx in [0, 2]:
    # Setting Region Variables
    print(f'Region {idx}: ')
    features = region[idx][['f0', 'f1', 'f2']]
    target = region[idx]['product']
    
    # Preforming the Cross-Validation
    scores = cross_val_score(models[idx], features, target, cv=5, scoring='neg_mean_squared_error')
    
    rmse = np.sqrt(-scores)
    print('RMSE for each fold:', rmse)
    print('Average RMSE:', np.mean(rmse))
```

<h4>LassoCV</h4>

```python
lasso_cv = LassoCV(alphas=ridge_alphas, cv=5, random_state=state)
for idx in [0, 2]:
    print(f'Region {idx}:')
    lasso_cv.fit(f_train[idx], t_train[idx])
    predictions = lasso_cv.predict(f_test[idx])
    rmse = np.sqrt(mean_squared_error(t_test[idx], predictions))
    print(f'Lasso RMSE: {rmse}')
```

<h2>Preperation for Profit Calculation</h2>
<p>&nbsp;&nbsp;&nbsp;&nbsp;Moving towards calculating the profit within each region. 
    </br>I'm going to: 
<ol>
    <li>store each value needed for profit calculation as a seperate variable</li>
    <li>calculate the volume of reserves needed to construct a new well without occuring loss</li>
    <li>compare that number to the average volume of each region</li>
</ol>

<h3>Preparing Variables and Defining Calculation Function</h3>

```python
# Creating individual lists to store the variables
top_200 = [] # For the top 200 sorted Locations
top_sum = [] # For each regions total sum of the top 200 locations
top_revenue = [] # For the revenue each of those regions will produce
profit = [] # for the total profit after building costs are removed.

def calculate_region_profit(locations, cost=100_000_000, price_per_thousand=4500, value_column='product', top=200):
    # Sorting locations on sort column
    sorted_locations = locations.sort_values(by=value_column, ascending=False)
    
    # Pulling the top locations after sorting
    top_locations = sorted_locations[:top] 
    
    # Summing those Values
    sum_of_values = top_locations[value_column].sum() 
    
    # Calculating Revenue
    revenue_from_sum = sum_of_values * price_per_thousand  
    
    # Calculating Profit
    profit = revenue_from_sum - cost 
    
    # Returning relevant variables in a practical order for future use
    return profit, top_locations, revenue_from_sum, sum_of_values 

print(f'Projected Profit\n')
for idx in range(3):
    # Assigning values needed for calculations
    gain, top_200_values, value_revenue, value_sum = calculate_region_profit(region[idx])
    
    # Storing relavent calculation variables
    top_200.append(top_200_values)
    top_sum.append(value_sum)
    top_revenue.append(value_revenue)
    profit.append(gain)
    
    print(f'Region {idx}: ${profit[idx]:,.2f}')
```

<h3>Finding the Break Even amount for Top Producing Wells in each region</h3>

```python
budget_per_well = 100_000_000 / 200
revenue_per_1000 = 4500
break_even = budget_per_well / revenue_per_1000

print(f'Break Even Amount for each region = {(break_even*200):,.2f}\nBreak Evan Amount for each Well: {break_even:,.2f}\n')
print(f'Average Volume of highest producing locations in each region.')

for idx in range(3):
    sorted_values = region[idx].sort_values(by='product', ascending = False)
    average = sorted_values['product'][:200].mean()

    # Compare with break-even volume
    comparison = "Sufficient" if average >= break_even else "Insufficient"
    
    print(f"Region {idx}: Average Volume = {average:,.2f}, ({comparison})")
```

<h3>Printing Calculation Results and Preparing for Bootstrapping Profits</h3>

```python
top_predictions = [] # Storing the top 200 predictions of each region
predicted_volume = [] # Storing the volume that is predicted to be there.
actual_values = [] # Storing the actual values based on samples
actual_volume = [] # Storing the actual total of reserves for all 200 top wells based on samples.
difference = [] # Storing the difference between the predicted and the sampled reserves

for idx in range(3):
    sorted_values = region[idx].sort_values(by='predicted_value', ascending = False)
    top_predictions.append(sorted_values[:200])
    top_indexes = sorted_values[:200].index
    
    predicted_volume.append(top_predictions[idx]['predicted_value'].sum())
    
    actual_values.append(region[idx].loc[top_indexes, 'product'])
    actual_volume.append(actual_values[idx].sum())
    
    difference.append(actual_volume[idx]-predicted_volume[idx])
    
    profit.append(calculate_region_profit(sorted_values))

    print(f"Region {idx}: Predicted Volume = {predicted_volume[idx]:,.2f}"
          f'\n\tActual Volume = {actual_volume[idx]:,.2f}'
          f'\n\tDifference : {abs(difference[idx]):,.2f} - Percentage: {(abs(difference[idx]/actual_volume[idx]) * 100):.2f}%'
          f'\n\n\tPredicted Profit: ${profit[idx]:,.2f}\n')

comparison = ((profit[2] - profit[0])/(profit[2]+profit[0]))*100
print(f'Difference between Region 0 and Region 2 Predicted Profit: {comparison:.2f}%')
```

<h2>Bootstraping for Profit reliability</h2>

```python
region_profit = []
region_mean = []

def bootstrap_profit(region_data, num_iterations=1000):
    profits = []
    
    for _ in range(num_iterations):
        sample = resample(region_data)
        profit,_,_,_  = calculate_region_profit(sample)
        profits.append(profit)
        
    avg_profit = np.mean(profits)
    
    # Calculating 95% Confidence Interval
    lower_ci = np.percentile(profits, 2.5)
    upper_ci = np.percentile(profits, 97.5)
    
    # Calculating Risk
    risk_of_loss = ((np.array(profits) < 0).mean() * 100)  # Expressed as a percentage
    
    
    return avg_profit, lower_ci, upper_ci, risk_of_loss

for idx in range(3): # Allowing me to check whatever regions I want
    sorted_values = region[idx].sort_values(by='product', ascending=False)[:200]
    print(f'Printing bootstraping results for region {idx}...')
    
    profit_average, low_ci, up_ci, risk = bootstrap_profit(sorted_values)
    region_profit.append(profit_average)
    
    print(f'\tProfit average with Bootstraping: ${profit_average:,.2f}'
          f'\n\t95% Cofidence Interval: (${low_ci:,.2f} | ${up_ci:,.2f})'
          f'\n\tRisk of Loss: {risk}%\n')
```
